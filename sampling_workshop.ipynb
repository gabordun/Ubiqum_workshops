```r
library(dplyr)
library(ggplot2)
```

```r
a<-data.frame(1:1000)
a
```

```r
set.seed(123)
b<-sample_n(a,0.2*nrow(a),replace=FALSE)
b
```

```r
c<-replicate(50,sample_n(a,0.2*nrow(a),replace=FALSE))
c
```

```r
d<-unlist(c)
d
```

```r
e<-unique(d)
e
```

```r
f<-as.data.frame(table(d))
f
```

```r
min(f[,2])
max(f[,2])
```
